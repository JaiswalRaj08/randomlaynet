```python
def create_batch(source, destination, batch_size=5000, num_batch = 1):
    images = sorted(os.listdir(source))
    
    for i in range(0, len(images), batch_size):
        batch_path = os.path.join(destination, 'batch_{}'.format(i+1))
        os.makedirs(batch_path, exist_ok=True)
        batch_files = images[i: i+batch_size]

        # Copy the files from the source folder to the batch folder
        for file_name in batch_files:
            source_file = os.path.join(source, file_name)
            destination_file = os.path.join(batch_path, file_name)
            shutil.copyfile(source_file, destination_file)

    print(f'Batch created.')
```

```python
source_dir = '/media/nas_mount/avinash_ocr/Rajeev/raj/avinash_sir_da/new-da-approach/new_df/publaynet/train'
destination_dir = '/media/nas_mount/avinash_ocr/Rajeev/raj/avinash_sir_da/new-da-approach/publaynet_batches'
create_batch(source=source_dir, destination=destination_dir)
```

```python
def plot_distribution(df):
    new_stats = dict(df['category_id'].value_counts())
    plt.bar(list(new_stats.keys()), list(new_stats.values()))

    class_label = list(new_stats.keys())
    class_count = list(new_stats.values())

    # Add labels to the top of each bar
    for i in range(len(class_label)):
        count = class_count[i]
        percentage = (count / sum(class_count)) * 100
        plt.text(class_label[i], count, f"{count} ({percentage:.2f}%)", ha='center', va='bottom')

    plt.xlabel('Category')
    plt.ylabel('Count')
    plt.title('Category Distribution')
    plt.show()
```
